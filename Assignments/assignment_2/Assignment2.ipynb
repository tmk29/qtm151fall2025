````python
import pandas as pd
import matplotlib.pyplot as plt

korea_emissions = pd.read_csv('data/korea_rep_co2_emissions.csv')
germany_emissions = pd.read_csv('data/germany_co2_emissions.csv')
columns = ["year", "total_emissions"]

ylbl = "CO2 Emissions (kilotons)"

x1 = [korea_emissions[columns[0]]]
y1 = [korea_emissions[columns[1]]]
plt.scatter(x1, y1)
plt.title("CO2 Emissions by Korea")
plt.xlabel("Year")
plt.ylabel("CO2 Emissions (kilotons)")
plt.show()

x2 = [germany_emissions[columns[0]]]
y2 = [germany_emissions[columns[1]]]
plt.scatter(x2, y2)
plt.title("CO2 Emissions by Germany")
plt.xlabel("Year")
plt.ylabel("CO2 Emissions (kilotons)")
plt.show()
````
